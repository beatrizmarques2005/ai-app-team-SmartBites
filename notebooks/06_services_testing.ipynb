```python
from recipe_crawler import RecipeCrawler
from recipe_scraper import RecipeScraper
import os

scraper = RecipeScraper()

recipe_urls = os.getenv("RECIPES_URLS", "").strip("()").split()

all_recipes = []

for site in recipe_urls:
    print(f"\n🔍 Crawling: {site}")

    crawler = RecipeCrawler(site)
    links = crawler.get_all_recipe_links()

    print(f"  ➤ Found {len(links)} recipe links")

    for link in links[:20]:  # limit while testing
        print(f"    📄 Extracting recipe from: {link}")
        recipe = scraper.scrape_recipe(link)

        if recipe:
            all_recipes.append(recipe)

print(f"\n🎉 Total recipes extracted: {len(all_recipes)}")

# Save them
import json
with open("recipes.json", "w", encoding="utf-8") as f:
    json.dump(all_recipes, f, indent=2, ensure_ascii=False)
```
