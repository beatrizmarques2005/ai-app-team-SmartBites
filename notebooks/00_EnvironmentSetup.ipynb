## <font color='#BFD72F' size=6>3.2 How We Secured the API Key</font> <a class="anchor" id="3_2"></a>

[Back to TOC](#toc)

To keep our credentials safe and avoid hardcoding them into the codebase:

- We created a `.env` file in the root of our project directory  
- Inside it, we added:

```env
GEMINI_API_KEY=your-api-key-here
```

## <font color='#BFD72F' size=6>3.3 Test the API Key</font> <a class="anchor" id="3_3"></a>

[Back to TOC](#toc)

````python
import os
from dotenv import load_dotenv
import google.generativeai as genai

# Load environment variables
load_dotenv()

# Configure the API
api_key = os.getenv("GOOGLE_API_KEY")
# api_key = 'AIzaSyDBc1fyta2yyu8VzEGZtTmuf0Vz736Pf8M' # not good practice to hardcode API keys in code. just as an example here.

if not api_key:
    print("❌ API key not found. Check your .env file!")
    print("Make sure you have GOOGLE_API_KEY=your_key in your .env file")
else:
    print("✅ API key found!")

    # Configure the generative AI
    genai.configure(api_key=api_key)

    # Create a model
    model = genai.GenerativeModel("gemini-2.5-flash")

    # Make your first API call!
    try:
        response = model.generate_content(
            "Hello! Please respond with a short, friendly greeting."
        )
        print("\n🎉 SUCCESS! Your first AI response:")
        print(f"{response.text}")
    except Exception as e:
        print(f"❌ Error making API call: {e}")
        print("Check your API key and internet connection")
````
