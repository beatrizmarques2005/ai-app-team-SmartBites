```python
from supabase import create_client
import os
from dotenv import load_dotenv

load_dotenv()

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")

print("Testing connection...")

try:
    supabase = create_client(SUPABASE_URL, SUPABASE_KEY)
    # Simple test query
    result = supabase.table('receipts').select('*').limit(1).execute()
    print("✅ Connection successful!")
    print("You can now use supabase in your notebook!")
except Exception as e:
    print("❌ Connection failed:", e)
```

```python
def insert_into_table(table_name, json_data):
    """
    Insert JSON data into any table
    
    Parameters:
    - table_name: name of the table to insert into
    - json_data: the JSON data to insert (Python dict)
    
    Returns:
    - The inserted record if successful, None if failed
    """
    try:
        result = supabase.table(table_name).insert(json_data).execute()
        
        if result.data:
            print(f"✅ Successfully inserted into '{table_name}' table!")
            print(f"   Record ID: {result.data[0].get('id', 'No ID field')}")
            return result.data[0]
        else:
            print(f"❌ Failed to insert into '{table_name}'")
            return None
            
    except Exception as e:
        print(f"❌ Error inserting into '{table_name}': {e}")
        return None
```
